# **Method 1 - TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#loading vectorizer and fitting all descriptions to it
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['Description'])
```

```python
import sklearn
def tfIDF(query,status = 1):
    print('Generated Description : ',query)
    query_vec = vectorizer.transform([query])
    
    #using cosine similarity to generate rank list
    results = sklearn.metrics.pairwise.cosine_similarity(X, query_vec, dense_output=True)
#     print(results.size)
    res = []
    for i in range(len(results)):
        res.append([results[i],i])
    
    res.sort(reverse = True)
    if status == 0:
        return res
    print('Recommended Video--------Recommendation')
    for i in res[:10]:
        print(df.iloc[i[1],0],"--------",df.iloc[i[1],1])
```

# **Graph Based Method**

```python
import networkx as nx
import tqdm.notebook as tqdm
from collections import defaultdict
import numpy as np
def cosine(a, b):
    return np.sum(a * b) / ((np.sum(a ** 2) ** 0.5) * (np.sum(b ** 2) ** 0.5))
```

```python
model = SentenceTransformer('all-mpnet-base-v2')
emb = model.encode(df.Description.values)
```

```python
from collections import defaultdict
def rec(test_sentence,status = 1):
    if status == 1:
        print("Query:", test_sentence)
    adjList = defaultdict(set)
    rel_sents = df.Description.values
    query_emb= model.encode(test_sentence)
    for i in tqdm.tqdm(range(len(rel_sents))):
        for j in range(i + 1, len(rel_sents)):
            if np.sum(emb[i]) and np.sum(emb[j]) and cosine(emb[i],emb[j]) >= 0.1:
                if cosine(emb[i], query_emb) >= 0.1 and cosine(emb[j], query_emb) >= 0.1:
                    adjList[i].add(j)
                    adjList[j].add(i)

                    
#graph construction
    G = nx.Graph()
    G.add_nodes_from(list(range(len(rel_sents))))

    for i in adjList.keys():
        for j in adjList[i]:
            G.add_edge(i, j)

            
#performing page-rank and obtaining pa
    pagerank = nx.pagerank(G)
    pr_scores = [0] * len(rel_sents)
    for key, val in pagerank.items():
        pr_scores[key] = val
#     pr_scores = np.array(pr_scores)
#     pr_scores /= np.sum(pr_scores ** 2) ** 0.5
    tmp = []
    for i in range(len(rel_sents)):
        tmp.append([pr_scores[i], i])
    tmp.sort(reverse = True)
    if status == 0:
        return tmp

    ranklist = []
    for [score, index] in tmp:
#         NetScore[index].append(score)
#         Norm3 += score
        ranklist.append(index)

    final_sents = []
    for i in ranklist[: 10]:
        final_sents.append([i,rel_sents[i]])
    print('Recommended Video--------Recommendation')
    for idx, sent in (final_sents):
        try:
            display(df[['VideoID', 'Description']].iloc[idx])
        except:
            print(df[['VideoID', 'Description']].iloc[idx])
```

# **Combining All 3 Methods**

```python

def normalise(l):
#     print(l)
    sum = 0
    for i in l:
        sum+=i[0]
    for i in range(len(l)):
        l[i][0] = l[i][0]/sum
    return l
        
# def def_value():
#     return 0.00

for query in queries:
#     print('Generated Description : ')
    list1  = normalise(tfIDF(query,0))
    list2  = normalise(generate_results(query,0))
    list3  = normalise(rec(query,0))
    
    D = collections.defaultdict(float)

    for [i,j] in list1:
        D[j]+=i[0]
    
    for [i,j] in list2:
        D[j]+=i[0]
        
    for [i,j] in list3:
        D[j]+=i
        
    res = []
    for i in D.keys():
        res.append([D[i],i])
    res.sort(reverse = True)
    
    print('Recommended Video--------Recommendation')
    
    for i in res[:10]:
        print(df.iloc[i[1],0],"--------",df.iloc[i[1],1])
    
    print('--------------------------------------------------------------------------------------')
    print('--------------------------------------------------------------------------------------')
        
    
    
    
    
    
```
